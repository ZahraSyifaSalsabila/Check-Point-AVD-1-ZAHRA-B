#Business Understanding




## Business Objective

Tujuan bisnis dari proyek ini adalah untuk memahami dan menganalisis biaya diet sehat di tingkat global dengan menggunakan dataset Global Healthy Diet Cost Analysis. Diet Sehat masuk dalam salah satu kebutuhan yang memiliki peran penting dalam meningkatkan kualitas hidup dan untuk mencegah berbagai penyakit. Tapi, pada kenyataannya tidak semua masyarakat di setiap negara memiliki akses yang mudah untuk memperoleh makanan bergizi karena tingginya harga pangan sehat. Karena itu, tujuan dibuatnya penelitian ini yaitu untuk mengetahui perbedaan biaya diet sehat antar negara kemudian mengamati perubahan tren dari waktu ke waktu.

## Assess Situation




 Melalui pembuatan Dataset Global Healthy Diet Cost Analysis, kita dapat memperoleh data tentang rata-rata biaya diet sehat dari berbagai negara untuk dapat menilai keterjangkauan diet sehat secara global. Namun, proses analisis ini memiliki beberapa tantangan misalnya ada kemungkinan data tidak lengkap untuk beberapa negara atau dalam beberapa periode tertentu, juga karena pengaruh eksternal misalnya seperti inflasi, konflik internasional, ataupun pandemi yang dapat memengaruhi harga bahan pangan namun ternyata tidak tercatat secara langsung dalam dataset.
