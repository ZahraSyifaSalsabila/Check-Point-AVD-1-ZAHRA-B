##**Outlier Values**


Di sini dilakukan pengecekan nilai ekstrem (outliers) pada kolom bertipe data numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

Dari hasil pengecekkan, diperoleh:
* Kolom country_code dan year

Persentase outliers = 0%

Tidak ada nilai ekstrem, artinya data pada kolom ini bersih dan konsisten.

* Kolom cost_healthy_diet_ppp_usd dan annual_cost_healthy_diet_usd

Persentase outliers â‰ˆ 1,81%

Terdapat sebagian kecil data yang termasuk nilai ekstrem. Ini menunjukkan bahwa ada beberapa negara atau tahun dengan biaya diet sehat yang cukup berbeda dibandingkan kebanyakan data. Perbedaan ini bisa dipengaruhi oleh kondisi ekonomi, inflasi, atau perbedaan metode perhitungan.

* Kolom cost_vegetables_ppp_usd dan cost_fruits_ppp_usd

Persentase outliers sangat rendah (0,29% dan 0,22%)

Sebagian besar data berada di dalam rentang yang normal. Nilai ekstrem hanya sedikit, jadi tidak memberikan pengaruh yang signifikan terhadap analisis.

* Kolom total_food_components_cost

Persentase outliers = 0,58%

Persentase ini masih tergolong rendah. Mayoritas data menunjukkan pola yang konsisten, sementara nilai ekstrem yang ada kemungkinan berasal dari kombinasi biaya komponen makanan tertentu yang tidak umum.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

##**Relationship/Hubungan**


Aktivitas: Menganalisis biaya diet sehat tiap komponen makanan (sayur, buah, total food components) per region menggunakan heatmap.

Tujuan: Mengetahui region mana yang menghabiskan biaya paling tinggi untuk masing-masing komponen makanan, sehingga bisa membandingkan pola diet antar region secara langsung.

Visualisasi: Heatmap

```python
plt.figure(figsize=(8,6))
sns.heatmap(df[['annual_cost_healthy_diet_usd','total_food_components_cost']].corr(),
            annot=True,
            cmap='viridis',
            fmt='.2f')
plt.title('Korelasi antara Annual Cost dan Total Food Components')
plt.show()
```

Berdasarkan heatmap korelasi, terlihat bahwa nilai korelasi antara annual_cost_healthy_diet_usd dan total_food_components_cost sebesar 0,69, yang menunjukkan hubungan positif yang cukup kuat. Artinya, semakin tinggi total biaya komponen makanan, maka semakin tinggi pula total biaya diet sehat per tahun di suatu negara atau region. Karena korelasi tidak mendekati 1, berarti masih ada faktor lain yang memengaruhi biaya diet sehat. Namun, total biaya komponen makanan tetap menjadi faktor utama yang menjelaskan perbedaan biaya antar negara atau region.
