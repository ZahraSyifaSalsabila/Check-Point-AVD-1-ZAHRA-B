##**Outlier Values**


Di sini dilakukan pengecekkan nilai ekstrem (outliers) pada kolom yang bertipe data kan numerik. Hal ini dilakukan agar mendapatkan insight yang akurat sehingga dapat memperkuat analisis dan model yang akan kita bangun nantinya.

Dari hasil pengecekkan, diperoleh:
* Kolom country_code dan year

Persentase outliers = 0%

Tidak ada nilai ekstrem, artinya data pada kolom ini bersih dan konsisten.

* Kolom cost_healthy_diet_ppp_usd dan annual_cost_healthy_diet_usd

Persentase outliers ≈ 1,81%

Hanya sebagian kecil data yang ekstrem. Ini menandakan sebagian kecil negara atau tahun memiliki biaya diet sehat yang jauh berbeda dari mayoritas. Bisa disebabkan oleh perbedaan ekonomi atau metodologi pengukuran.

* Kolom cost_vegetables_ppp_usd dan cost_fruits_ppp_usd

Persentase outliers sangat rendah (0,29% dan 0,22%)

Hampir seluruh data berada dalam rentang normal. Outlier hampir tidak signifikan, jadi tidak menjadi masalah besar untuk analisis.

* Kolom total_food_components_cost

Persentase outliers = 0,58%

Masih rendah, menandakan mayoritas total biaya komponen makanan relatif konsisten. Outlier sedikit, mungkin berasal dari kombinasi biaya sayur, buah, dan komponen lainnya yang tidak biasa.

```python
results = []

cols = df.select_dtypes(include=['float64', 'int64'])

for col in cols:
  q1 = df[col].quantile(0.25)
  q3 = df[col].quantile(0.75)
  iqr = q3 - q1
  lower_bound = q1 - 1.5*iqr
  upper_bound = q3 + 1.5*iqr
  outliers = df[(df[col] < lower_bound) | (df[col] > upper_bound)]
  percent_outliers = (len(outliers)/len(df))*100
  results.append({'Kolom': col, 'Persentase Outliers': percent_outliers})

# Dataframe dari list hasil
results_df = pd.DataFrame(results)
results_df.set_index('Kolom', inplace=True)
results_df = results_df.rename_axis(None, axis=0).rename_axis('Kolom', axis=1)

# Tampilkan dataframe
display(results_df)
```

# **Construct Data**

Di Construct Data saya membuat kolom baru yaitu cost_level. Pengelompokan ini ditentukan dari besarnya biaya diet sehat per hari (dalam PPP USD). Dengan mengelompokkan negara ke dalam kategori seperti rendah, sedang, dan tinggi, pola distribusi biaya dapat terlihat lebih jelas dibandingkan jika hanya menggunakan angka kontinu.

Adapun tujuan pembuatan kolom baru ini yaitu:
1. Mempermudah perbandingan antar wilayah.
2. Mengidentifikasi kelompok negara dengan beban biaya diet tinggi.

```python
def kategori_biaya(cost_healthy_diet_ppp_usd):
    if cost_healthy_diet_ppp_usd > 3.5:
        return 'High'
    elif cost_healthy_diet_ppp_usd < 2:
        return 'Low'
    else:
        return 'Medium'

df['cost_level'] = df['cost_healthy_diet_ppp_usd'].apply(kategori_biaya)
```
